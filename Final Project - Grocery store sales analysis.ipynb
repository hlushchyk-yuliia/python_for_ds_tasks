**`У файлі деякі показники розраховані невірно (Net Bill Amount, Gross Bill Amount, % Profit Margin, Profit Margin), 
тому для аналізу буду використовувати правильні розрахунки:`**

```python
df.drop(columns=['Net Bill Amount'], inplace=True)

print(df)

net_bill_amount_by_receipt = df.groupby("Receipt Number")["Amount"].sum().reset_index()

df = df.merge(net_bill_amount_by_receipt, on="Receipt Number", suffixes=('', '_sum'))

df.rename(columns={"Amount_sum": "Net Bill Amount"}, inplace=True)

df["Gross Bill Amount"] = df["Net Bill Amount"] * df["GST"]

df["Profit Margin"] = df["Amount"] - df["Operating Cost"] - df["Product Cost"]

df["% Profit Margin"] = (df["Profit Margin"] / df["Amount"])
```

```python
# додамо бібліотеки, які будемо використовувати
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm, pearsonr, stats
```

1. Як змінюється обсяг продажів і прибутку в розрізі кварталів?

```python
df['Quarter'] = (df['Month'] - 1) // 3 + 1

quarterly_profit_margin = df.groupby(['Year', 'Quarter'])['Profit Margin'].mean().reset_index()
quarterly_sales = df.groupby(['Year', 'Quarter'])['Amount'].sum().reset_index()

plt.figure(figsize=(18, 6))

plt.plot(quarterly_sales['Year'].astype(str) + '-Q' + quarterly_sales['Quarter'].astype(str), 
         quarterly_sales['Amount'], marker='o', label='Обсяг продажів', color='b')

plt.twinx()
plt.plot(quarterly_profit_margin['Year'].astype(str) + '-Q' + quarterly_profit_margin['Quarter'].astype(str), 
         quarterly_profit_margin['Profit Margin'], marker='o', label='Прибуток', color='g')

plt.xlabel('Рік-Квартал')
plt.ylabel('Сума')
plt.title('Зміна обсягу продажів та прибутку в розрізі кварталів')

plt.grid()

plt.xticks(rotation=90, ha='center', va='top')

plt.text(0.02, 0.97, 'Обсяг продажів', color='b', transform=plt.gca().transAxes, fontsize=12, fontweight='bold')
plt.text(0.02, 0.9, 'Прибуток', color='g', transform=plt.gca().transAxes, fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

Висновок: Графіки, які відображають зміну обсягу продажів та прибутку в розрізі кварталів протягом року, дають нам важливий інсайт щодо фінансової динаміки підприємства. За цими графіками видно, що обсяг продажів поступово зростає протягом року. Це може свідчити про позитивний тренд у розвитку бізнесу та збільшення популярності продукції серед споживачів.

У той же час, прибуток виявляє значні коливання в різні квартали. Ці коливання можуть бути результатом різних факторів, таких як зміни в сезонності попиту, конкурентна ситуація на ринку, зміни в витратних складових виробництва, і т. д. Важливо звертати увагу на ці коливання та аналізувати їх для покращення фінансової стабільності підприємства.

В цілому, графіки демонструють, що, хоча обсяги продажів зростають, ефективність управління прибутком потребує уваги та може бути поліпшена шляхом більшої стабільності та управління витратами.

17. Проведення A/B-тесту на вплив географічного розташування на обсяги продажів.

```python
group_a = df[(df['Country'].isin(['Indonesia', 'Malaysia']))]
group_b = df[(df['Country'].isin(['Philippines', 'Singapore']))]

t_stat, p_value = stats.ttest_ind(group_a['Amount'], group_b['Amount'], equal_var=False)

alpha = 0.05

print("t-Statistic:", t_stat)
print("p-Value:", p_value)

if p_value < alpha:
    print("Статистично значуща різниця: вплив географічного розташування є.")
else:
    print("Різниця не є статистично значущою: географічне розташування не впливає на обсяги продажів.")
```

18. ABC-аналіз згідно груп:
- Група A - позиції, що приносять найбільший прибуток (80% прибутку від продажів 20% асортименту).
- Група B - товари зі стабільно середнім попитом.
- Група C - категорія товарів із найменшою рентабельністю.

```python
df.sort_values(by='Profit Margin', ascending=False, inplace=True)

df['Cumulative Profit'] = df['Profit Margin'].cumsum()

total_profit = df['Profit Margin'].sum()

threshold_a = 0.8 * total_profit

threshold_b = 0.2 * total_profit + threshold_a

df['ABC Category'] = 'C'
df.loc[df['Cumulative Profit'] <= threshold_a, 'ABC Category'] = 'A'
df.loc[(df['Cumulative Profit'] > threshold_a) & (df['Cumulative Profit'] <= threshold_b), 'ABC Category'] = 'B'

print(df[['Items', 'Profit Margin', 'ABC Category']])
```

```python
abc_counts = df['ABC Category'].value_counts()

plt.bar(abc_counts.index, abc_counts.values, label='Number of Products')
plt.xlabel('ABC Category')
plt.ylabel('Number of Products')
plt.title('ABC Analysis')
plt.legend()
plt.show()
```
